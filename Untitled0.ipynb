```python
# AI-Driven Customer Preference Prediction
# Hamna's Mini Project for MBZUAI Business Stream
# Python 3.x

import pandas as pd
from sklearn.tree import DecisionTreeClassifier
import matplotlib.pyplot as plt

# ------------------------------
# 1. Dataset (you can add more rows)
# ------------------------------
data = {
    'CustomerID': ['C001','C002','C003','C004','C005','C006','C007','C008'],
    'Age':[22,30,27,21,35,28,24,32],
    'Gender':['F','M','F','M','F','F','M','M'],
    'LastPurchase':['Electronics','Clothing','Electronics','Grocery','Clothing','Electronics','Clothing','Grocery'],
    'PurchaseFreq':[3,2,4,1,2,3,2,1],
    'Feedback':[4,3,5,2,4,5,3,2]
}

df = pd.DataFrame(data)

# ------------------------------
# 2. Encode categorical data
# ------------------------------
df['Gender_enc'] = df['Gender'].map({'F':0,'M':1})
df['Category_enc'] = df['LastPurchase'].map({'Electronics':0,'Clothing':1,'Grocery':2})

# Features and target
X = df[['Age','Gender_enc','PurchaseFreq','Feedback']]
y = df['Category_enc']

# ------------------------------
# 3. Train Decision Tree
# ------------------------------
model = DecisionTreeClassifier()
model.fit(X, y)

# Predict category for all customers
df['Predicted_enc'] = model.predict(X)

# Map back to category names
category_map = {0:'Electronics',1:'Clothing',2:'Grocery'}
df['PredictedCategory'] = df['Predicted_enc'].map(category_map)

# ------------------------------
# 4. Output prediction table
# ------------------------------
print("Customer Preference Prediction:")
print(df[['CustomerID','LastPurchase','PredictedCategory']])

# ------------------------------
# 5. Simple Charts
# ------------------------------
# Most purchased category
df['LastPurchase'].value_counts().plot(kind='bar', color='skyblue')
plt.title('Most Purchased Categories')
plt.xlabel('Category')
plt.ylabel('Count')
plt.show()

# Purchase frequency vs Feedback
plt.scatter(df['PurchaseFreq'], df['Feedback'], color='green')
plt.title('Purchase Frequency vs Feedback')
plt.xlabel('Purchase Frequency')
plt.ylabel('Feedback Score')
plt.show()

# ------------------------------
# 6. Business Recommendation
# ------------------------------
recommendation = """
Business Recommendation:
- Focus marketing on categories with highest predicted purchases (Electronics).
- Offer loyalty programs to frequent purchasers to increase retention.
- Monitor customer feedback to improve service quality.
Ethical Note:
- Ensure customer data privacy and avoid biases in decision-making.
"""
print(recommendation)
```
